Getting info for books in a web page.

```python
product_list = []

for product in soup.find_all("article", class_="product_pod"):
    price = product.find('p', class_="price_color").text
    title = product.find('h3').find("a").get("title")
    book_url = product.find('h3').find("a").get("href")
    rating = product.find('p', class_="star-rating").get("class")[-1]
    availability = product.find("p", class_="instock").text.strip()

    product_dict = {
        "price": price,
        "title": title,
        "book_url": book_url,
        "rating": rating,
        "availability": availability
    }

    product_list.append(product_dict)
```

```python
pd.DataFrame(product_list)
```

```python
def scrape_page(url: str) -> pd.DataFrame:
    response = requests.get(url)
    soup = BeautifulSoup(response.text)

    product_list = []

    for product in soup.find_all("article", class_="product_pod"):
        price = product.find('p', class_="price_color").text
        title = product.find('h3').find("a").get("title")
        book_url = product.find('h3').find("a").get("href")
        rating = product.find('p', class_="star-rating").get("class")[-1]
        availability = product.find("p", class_="instock").text.strip()

        product_dict = {
            "price": price,
            "title": title,
            "book_url": book_url,
            "rating": rating,
            "availability": availability
        }

        product_list.append(product_dict)

    return pd.DataFrame(product_list)
```

```python
scrape_page("https://books.toscrape.com/catalogue/category/books/travel_2/index.html")
```

```python
def get_page_no(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text)
    try:
        no_pages = soup.find_all('ul', class_="pager")[0].text.replace('next', '').strip().split("of ")[-1]
    except IndexError as e:
        print(e)
        no_pages = 1
    return int(no_pages)
```
